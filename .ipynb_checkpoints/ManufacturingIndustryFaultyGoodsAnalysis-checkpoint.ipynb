# 4번 그룹에서 타 그룹 대비 불량품의 비율이 타 그룹 대비 양품의 비율보다 조금 증가하지만, 1. 그 차이가 작으며 2. 견학 당시 도메인 전문가가 워크 그룹 간의 차이에 대해 설명하지 않았음, 앞의 두 이유로 해당 파라미터는 생략함

# 데이터 분석으로 선택된 파라미터는 PT_WIDTH, PT_LENGTH, FUR_SZ_TEMP, FUR_EXTEMP, ROLLING_TEMP_T5, HSB이며 데이터는 ROLLING_TEMP_T5 < 100의 이상치를 제거하여 사용한다.

```python
df_parsed = pd.get_dummies(df[df['ROLLING_TEMP_T5'] > 100][['PT_WDTH', 'PT_LTH', 'FUR_SZ_TEMP', 'FUR_EXTEMP', 'ROLLING_TEMP_T5', 'HSB']])
```

```python
df_y = df[df['ROLLING_TEMP_T5'] > 100]['SCALE']
len(df_y)
```

# SVM 등 특별한 모델이 아닐시에는 굳이 스케일을 해줄 필요는 없지만, 1. SVM을 곧 쓸 것이기 때문에, 2. 스케일을 해도 되기 때문에, 이상 두 가지 이유로 스케일을 한다.

```python
from sklearn.preprocessing import StandardScaler
```

```python
feature_names = df_parsed.columns
scaler = StandardScaler()
df_parsed  = scaler.fit_transform(df_parsed)
df_parsed = pd.DataFrame(df_parsed, columns = feature_names)

df_parsed.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_train_x, df_test_x, df_train_y, df_test_y = train_test_split(df_parsed, df_y, test_size = 0.3, random_state = 777)
```

```python
df_train_x, df_validation_x, df_train_y, df_validation_y = train_test_split(df_train_x, df_train_y, test_size = 3/7, random_state = 777)
```

# Decision Tree 모델로 정확도가 얼마나 나오는지, 파라미터를 어떻게 사용하여 데이터를 분류하는지 확인한다.

```python
train_accuracy = list()
test_accuracy = list()

para_leaf = [leaf for leaf in range(1, 21)]

for leaf in para_leaf:
    tree = DecisionTreeClassifier(random_state = 777, min_samples_leaf = leaf)
    tree.fit(df_train_x, df_train_y)
    train_accuracy.append(tree.score(df_train_x, df_train_y))
    test_accuracy.append(tree.score(df_validation_x, df_validation_y))

score_df = pd.DataFrame()
score_df['leaf'] = para_leaf
score_df['train'] = train_accuracy
score_df['test'] = test_accuracy
score_df.round(3)
```

```python
plt.plot(para_leaf, train_accuracy, linestyle = '-', label = 'train')
plt.plot(para_leaf, test_accuracy, linestyle = '--', label = 'test')
plt.xlabel('leaf')
plt.ylabel('accuracy')
plt.legend()
```

# 1. train > test 2. |train - test|가 작을 것 3. 변곡점을 만나기 전 4. 과적합을 줄이기 위해서 최대한 많은 수의 '잎사귀 노드 내의 데이터 수' 선택 4가지 조건을 모두 만족하는 leaf는 8

```python
train_accuracy = list()
test_accuracy = list()

para_split = [split * 2 for split in range(1, 21)]

for split in para_split:
    tree = DecisionTreeClassifier(random_state = 777, min_samples_split = split, min_samples_leaf = 8)
    tree.fit(df_train_x, df_train_y)
    train_accuracy.append(tree.score(df_train_x, df_train_y))
    test_accuracy.append(tree.score(df_validation_x, df_validation_y))

score_df = pd.DataFrame()
score_df['split'] = para_split
score_df['train'] = train_accuracy
score_df['test'] = test_accuracy
score_df.round(3)
```

# leaf = 8일때 min_samples_split은 데이터에 영향을 주지 않는 것으로 보인다. 따라서 split은 따로 설정하지 않는다.

```python
train_accuracy = list()
test_accuracy = list()

para_depth = [depth for depth in range(1, 16)]

for depth in para_depth:
    tree = DecisionTreeClassifier(random_state = 777, max_depth = depth, min_samples_leaf = 8)
    tree.fit(df_train_x, df_train_y)
    train_accuracy.append(tree.score(df_train_x, df_train_y))
    test_accuracy.append(tree.score(df_validation_x, df_validation_y))

score_df = pd.DataFrame()
score_df['depth'] = para_depth
score_df['train'] = train_accuracy
score_df['test'] = test_accuracy
score_df.round(3)
```

```python
plt.plot(para_depth, train_accuracy, linestyle = '-', label = 'train')
plt.plot(para_depth, test_accuracy, linestyle = '--', label = 'test')
plt.xlabel('depth')
plt.ylabel('accuracy')
plt.legend()
```

# depth가 낮을 때 test가 train보다 높은 값을 보인다. 상식적으로 이해가 가지 않는 부분이지만, 일단 test < train인 부분에서 가장 적은 (과적합을 피하기 위해서) 5를 선택한다. max_depth == 5

```python
tree_fin = DecisionTreeClassifier(max_depth=5, min_samples_leaf=8, random_state = 777)
tree_fin.fit(df_train_x, df_train_y)
print(f'Accuracy on training set: {tree_fin.score(df_train_x, df_train_y)}')
print(f'Accuracy on test set: {tree_fin.score(df_test_x, df_test_y)}')
```

# DecisionTreeClassifier로 만든 모델의 정확도는 0.944로 확인되며 가장 상위에 위치한 파라미터는 ROLLING_TEMP_T5이다.

```python
train_accuracy = list()
test_accuracy = list()

para_estimator = [estimator * 10 for estimator in range(1, 16)]

for estimator in para_estimator:
    tree = RandomForestClassifier(random_state = 777, max_depth = 5, min_samples_leaf = 8, n_estimators=estimator)
    tree.fit(df_train_x, df_train_y)
    train_accuracy.append(tree.score(df_train_x, df_train_y))
    test_accuracy.append(tree.score(df_validation_x, df_validation_y))

score_df = pd.DataFrame()
score_df['estimator'] = para_estimator
score_df['train'] = train_accuracy
score_df['test'] = test_accuracy
score_df.round(3)
```

```python
plt.plot(para_estimator, train_accuracy, linestyle = '-', label = 'train')
plt.plot(para_estimator, test_accuracy, linestyle = '--', label = 'test')
plt.xlabel('estimator')
plt.ylabel('accuracy')
plt.legend()
```
