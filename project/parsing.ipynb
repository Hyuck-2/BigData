```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import random
```

```python
df11 = pd.read_csv('df11.csv')
df12 = pd.read_csv('df12.csv')
df21 = pd.read_csv('df21.csv')
df22 = pd.read_csv('df22.csv')
```

```python
wine_group = pd.read_csv('wine_raw_data_group.csv', encoding = 'cp949')
```

```python
wine_group
```

```python
A = wine_group[wine_group['그룹'] == 'A'].index[0]
B = wine_group[wine_group['그룹'] == 'B'].index[0]
C = wine_group[wine_group['그룹'] == 'C'].index[0]
D = wine_group[wine_group['그룹'] == 'D'].index[0]
E = wine_group[wine_group['그룹'] == 'E'].index[0]
```

```python
df11.head()
```

```python
dc = df11.copy()
dc.head()
```

```python
tmp_list = list()

for idx, group in enumerate(dc['choice1'].unique()):
    tmp_list.append(dc[dc['choice1'] == group])

    for cat in [category for category in wine_group.columns if category != '그룹']:
        tmp_list[idx]['choice1_'+cat] = wine_group[wine_group['그룹'] == group].loc[wine_group[wine_group['그룹'] == group].index[0]][cat]

len(tmp_list)
```

```python
result = pd.concat(tmp_list)
result.sort_index(inplace = True)
result.head()
```

```python
tmp_list = list()

for idx, group in enumerate(dc['choice2'].unique()):
    tmp_list.append(result[result['choice2'] == group])

    for cat in [category for category in wine_group.columns if category != '그룹']:
        tmp_list[idx]['choice2_'+cat] = wine_group[wine_group['그룹'] == group].loc[wine_group[wine_group['그룹'] == group].index[0]][cat]

result = pd.concat(tmp_list)
result.sort_index(inplace = True)
result.head()
```

```python
result.columns
```

```python
result.to_csv('df11_new.csv', index = False)
```
